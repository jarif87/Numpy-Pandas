# sorting

```python
sort_df=df.sort_values(by="Name")
sort_df.head()
```

```python
sort_df=df.sort_values(by="Name",ascending=False)
sort_df.head()
```

```python
sort_df=df.sort_values(by=["Name","Country"],ascending=True)
sort_df.head()
```

```python
sort_df=df.sort_values(by=["Country","Name"],ascending=True)
sort_df.sort_index(inplace=True)
sort_df.head()
```

```python
df_concat=pd.concat([df,sort_df])
df_concat.head()
```

```python
df_concat_key=pd.concat([df,sort_df],keys=["normal","sorted"])
df_concat_key.head()
```

```python
df_concat_key.tail()
```

```python
df_concat_key.loc["normal"]
```

```python
df_concat_key.loc["sorted"]
```
